```python
import os
import shutil

# ----- CHANGE THIS PATH -----
base_dir = r"D:\Minor Project Related\Data Sets ( Oily, Dry , Normal )\ALL\archive_1(Splitted Data)\Oily-Dry-Skin-Types"   # <- put your dataset path here
output_dir = r"D:\Minor Project Related\Data Sets ( Oily, Dry , Normal )\ALL\archive_1"
# ----------------------------

splits = ["train", "test", "val"]
classes = ["dry", "normal", "oily"]

# Create output folders
for cls in classes:
    os.makedirs(os.path.join(output_dir, cls.capitalize()), exist_ok=True)

# Move images
for split in splits:
    for cls in classes:
        src = os.path.join(base_dir, split, cls)
        dest = os.path.join(output_dir, cls.capitalize())

        if os.path.exists(src):
            for img in os.listdir(src):
                img_path = os.path.join(src, img)
                if os.path.isfile(img_path):
                    shutil.copy(img_path, dest)

print("✅ Dataset successfully combined into Dry, Normal, and Oily!")
```

## Only use this to rename and combine all folders

```python
import os
import shutil

SOURCE_ROOT = r"D:\Minor Project Related\Data Sets ( Oily, Dry , Normal )\ALL\Final Work\original_dataset"
DEST_ROOT = r"D:\Minor Project Related\Data Sets ( Oily, Dry , Normal )\ALL\Final Work\combined_dataset"

CLASSES = ["dry", "normal", "oily"]
VALID_EXTS = ('.jpg', '.jpeg', '.png')

# Create destination folders
for cls in CLASSES:
    os.makedirs(os.path.join(DEST_ROOT, cls), exist_ok=True)

# Counters for each class
counters = {cls: 1 for cls in CLASSES}

for batch in sorted(os.listdir(SOURCE_ROOT)):
    batch_path = os.path.join(SOURCE_ROOT, batch)

    if not os.path.isdir(batch_path):
        continue

    print(f"Processing: {batch}")

    for cls in CLASSES:
        class_src = os.path.join(batch_path, cls)

        if not os.path.exists(class_src):
            continue

        class_dst = os.path.join(DEST_ROOT, cls)

        for img in sorted(os.listdir(class_src)):
            if not img.lower().endswith(VALID_EXTS):
                continue

            src_img_path = os.path.join(class_src, img)
            ext = os.path.splitext(img)[1]

            new_name = f"{cls}_{batch}_{counters[cls]:05d}{ext}"
            dst_img_path = os.path.join(class_dst, new_name)

            shutil.copy2(src_img_path, dst_img_path)
            counters[cls] += 1

print("\n✅ All images combined and renamed successfully!")
print("Final counts:", counters)

```
