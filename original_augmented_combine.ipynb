## Augument + Original lai combine garney 


```python
import os
import shutil
```

```python
original_dir = r"D:\Minor Project Related\Data Sets ( Oily, Dry , Normal )\ALL\Final Work\combined_dataset_split\train"
augmented_dir = r"D:\Minor Project Related\Data Sets ( Oily, Dry , Normal )\ALL\Final Work\augmented_dataset_only_train"
combined_dir = r"D:\Minor Project Related\Data Sets ( Oily, Dry , Normal )\ALL\Final Work\train_augment+original"

os.makedirs(combined_dir, exist_ok=True)

classes = ["Dry", "Normal", "Oily"]
```

```python
for cls in classes:
    os.makedirs(os.path.join(combined_dir, cls), exist_ok=True)

    # Copy original images
    for img in os.listdir(os.path.join(original_dir, cls)):
        src = os.path.join(original_dir, cls, img)
        dst = os.path.join(combined_dir, cls, f"orig_{img}")
        shutil.copy(src, dst)

    # Copy augmented images
    for img in os.listdir(os.path.join(augmented_dir, cls)):
        src = os.path.join(augmented_dir, cls, img)
        dst = os.path.join(combined_dir, cls, f"aug_{img}")
        shutil.copy(src, dst)

print("✅ Original and augmented datasets combined successfully.")
```
